```python
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv("raw_data/A20521480_CollectedData_annotation.csv")
```

```python
# Take a first look to the dataset
df.head()
```

```python
# Set the file name as the index and rename its header
df = df.set_index("Unnamed: 2")
df.index.name = "file_name"

# Delete the first two columns which do not provide useful info
df = df.iloc[:, 2:]

# Create a list with the first three rows concatenated to modify the headers
columns = []
for i in range(df.shape[1]):
    columns.append("_".join(list(df.iloc[:3, i])))

# Delete the first three rows
df = df.iloc[3:, :]

# Rename the columns
df.columns = columns
```
