# Functions


```python
def read_file(file):
  """"
  Reads a data file and returns the dataframe containing the data
  :param file: name of the file 
  """
  if file[-3:] == "csv":
    df = pd.read_csv(annotations_dir+file)
  else:
    df = pd.read_excel(annotations_dir+file)

  df = df.replace(r'^\s*$', np.nan, regex=True)
  return df.fillna(1e-5)

def fill_annotation(row):
  """"
  Function that returns the annotation dictionary of a row in the dataframe
  :param row: row of a dataframe directly imported of the csv file
  """

  topleft = {"x":float(row[3]),
            "y":float(row[4])}
  rightdown = {"x":float(row[5]),
            "y":float(row[6])}

  height = abs(topleft["y"]-rightdown["y"])
  width = abs(topleft["x"]-rightdown["x"])

  topleft2 = {"x":float(row[21]),
            "y":float(row[22])}
  rightdown2 = {"x":float(row[23]),
            "y":float(row[24])}

  height2 = abs(topleft2["y"]-rightdown2["y"])
  width2 = abs(topleft2["x"]-rightdown2["x"])

  annotation_dict = {
  'annotation': [
  {'mouse1': 'boundingBox',
  'height': height,
  'width': width,
  'x': width/2+min(topleft["x"],rightdown["x"]),
  'y': height/2+min(topleft["y"],rightdown["y"])},
  {'nose': 'point', 'x': float(row[7]), 'y': float(row[8])},
  {'leftear': 'point', 'x': float(row[9]), 'y': float(row[10])},
  {'rightear': 'point', 'x': float(row[11]), 'y': float(row[12])},
  {'leftHip': 'point', 'x': float(row[13]), 'y': float(row[14])},
  {'rightHip': 'point', 'x': float(row[15]), 'y': float(row[16])},
  {'tailBase': 'point', 'x': float(row[17]), 'y': float(row[18])},
  {'TailEnd': 'point', 'x': float(row[19]), 'y': float(row[20])},



  {'mouse1': 'boundingBox',
  'height': height2,
  'width': width2,
  'x': width2/2+min(topleft2["x"],rightdown2["x"]),
  'y': height2/2+min(topleft2["y"],rightdown2["y"])},
  {'nose': 'point', 'x': float(row[25]), 'y': float(row[26])},
  {'leftear': 'point', 'x': float(row[27]), 'y': float(row[28])},
  {'rightear': 'point', 'x': float(row[29]), 'y': float(row[30])},
  {'leftHip': 'point', 'x': float(row[31]), 'y': float(row[32])},
  {'rightHip': 'point', 'x': float(row[33]), 'y': float(row[34])},
  {'tailBase': 'point', 'x': float(row[35]), 'y': float(row[36])},
  {'TailEnd': 'point', 'x': float(row[37]), 'y': float(row[38])},
  ],

  'class': 'image',
  'filename': row[2]}


  return annotation_dict


def process_df(df):
  data_df = df.iloc[4:]

  return [fill_annotation(data_df.iloc[i,:]) for i in range(len(data_df))]
```

# Format annotations

```python
frames =  "/content/gdrive/MyDrive/Colab Notebooks/TFM/Master_Mouse_Project/Data/Mouse_behavior_video_recording/frames/"
annotations_dir ="/content/gdrive/MyDrive/Colab Notebooks/TFM/Master_Mouse_Project/Data/Mouse_behavior_video_recording/annotation_result/"


# Set the annotations files 
annotations_files = os.listdir(annotations_dir)

# I had to delete this file due to problems when importing the data
annotations_files.remove("A20495939_CollectedData_annotation - CollectedData_annotation.csv")
annotations_list = []
for file in annotations_files:
  annotation = read_file(file)
  annotations_list.append(process_df(annotation)[0])


# saving in json format
with open("annotations.json", "w") as outfile:
    json.dump(annotations_list, outfile)
```
